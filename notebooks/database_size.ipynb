```python
import pandas as pd
import sqlite3

def get_size():
    query = f"""
                SELECT
                  SUM(pgsize) bytes,
                  name
                FROM dbstat
                GROUP BY name
                ORDER BY bytes DESC
              """
    return pd.read_sql_query(query, 'sqlite:////config/home-assistant_v2.db')
```

```python
%%time
size = get_size()
size
```
